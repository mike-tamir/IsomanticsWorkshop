# SVD Heatmaps

### Importing libraries

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns; sns.set()
import pandas as pd

# Set languages
languages = ['en','ru','de','es','fr','it', 'zh-CN']

# Set cmaps for color scale
cmap = sns.diverging_palette(10, 130, sep=20, as_cmap=True)
cmap1 = sns.diverging_palette(130, 10, sep=20, as_cmap=True)


```

```python
s_min_df = pd.read_csv("../HeatmapData/T/s_min.csv")
s_min_df = s_min_df.set_index([languages])
s_min_hm = sns.heatmap(s_min_df,cmap=cmap)
title = s_min_hm.set_title('s_min')
```
